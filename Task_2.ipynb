# Modified data

```python
modified_dfmodified_df=data.dropna().to_csv('modified_data',index=False)
df=pd.read_csv('modified_data')
df
```

```python
df.columns
```

```python
df.columns=['Region','Date','Frequency','EstimatedUnemployedRate(%)','EstimatedEmployed','EstimatedLabourParticipatedRate(%)','Area']
df['Date']=pd.to_datetime(df['Date'],dayfirst=True)
df['Month']=df['Date'].dt.month_name()
df['Year']=df['Date'].dt.year
df
```

```python
print("NULL VALUES IN EACH COLUMNS AFTER MODIFYING DATA")
df.isnull().sum()
```

```python
date_emp=df.groupby(['Year'])[['EstimatedUnemployedRate(%)','EstimatedEmployed','EstimatedLabourParticipatedRate(%)']].mean()
date_emp.round(2)
```

```python
region=df.groupby(['Region'])[['EstimatedUnemployedRate(%)','EstimatedEmployed','EstimatedLabourParticipatedRate(%)']].mean()
region.round(2)
```
